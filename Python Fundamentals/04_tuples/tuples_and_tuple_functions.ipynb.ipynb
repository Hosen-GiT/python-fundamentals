# Create a tuple

```python
# Create a tuple with a single item in it
t2= (2)
print(t2)
print(type(t2))

# Wrong way to write single item tuple
# Result shows only 2(integer) without parenthesis()
# This isn't an example of a single item tuple. It jsut an integer item


# The above example is not actually a single item tuple
# because it doesn't have any parenthesis()
# Tuple always contain parenthesis ()

# So the real example of a single item tuple is to add a comma(,) just after the item
t1= ('Akter',)
print(t1)

# The result give real single item tuple with required parenthesis () with ('Akter',)
```

# How to access to items in a tuple

1. Indexing
2. Slicing

```python
# Access item usng index, positive indexing left to right
# Accessing/Fetcing single at a time
t3= (1,2,3,4)
  #(0,1,2,3), postive index
#(-4,-3,-2,-1), negative index
print(t3[0])
print(t3[2])


# Negative indexing right to left
print(t3[-2])
print(t3[-1])
```

```python
# Slicing tuples with postive and negative slicing
# Accessing/fetching multiple items at a time
# Positive slicing
T4= (1,2,3,4,5,6)
print(T4[0:4])
# we are printing here index 0 to 4, the indexing starts from 0 and stops at index 3
T4= (1,2,3,4,5,6)
print(T4[0:4:2]) # Reverse print of 4,5,6

# Negative indexing with slicing
print(T4[-3:-1])
print(T4[-1:-4:-1])

# Reversing tuple
print(T4[::-1])


```

```python
# Fetching or extracting the items from tuple
# Example
T1= (1,2,3,(4,5,6))
print(T1[-1])
print(T1[-1][0])

# The code works on sub-tuple (4,5,6)
```

# How to edit items in Tuples

- So we cannot execute addition(append), edit and del(deletion) statement on tuple as tuple is immutable
- Only we can do to delete whole tuple using del(delete) statement

```python
# Edinting tuples or changing
T4= (1,2,3,4,5,6)
T4[0]= 7
print(T4)

# You see an error beacuse as the definition of tuple, tuples are immutable just like string
```

```python
# Adding items in tuple

T4= (1,2,3,4,5,6)
T4.append(7)

# Error
# Because we cannot add anythong once tuple is created as it is immutable
# Addition is not possible
```

```python
# Delete an item from a tuple
# We cannot delete any item from tuple individually as it contradicts the tuple's immutable property
# Cannot delete any portion
T4= (1,2,3,4,5,6)
print(T4) # We print T4

#del T4[0] # Not possible

# What we can do, either delete whole tuple and this is only possible
del T4 # We delete whole tuple T4 and in result T4 is not defined because it is deleted
print(T4)

```

```python
# Deleting a portion from 2D tuble

T1= (1,2,3,(4,5,6))
del T1[-1]

# Error because tuple is immutable
```

# Distinguishes between Lists and Tuples
- Syntax: tuple brace () and list has []
- Mutability: lsit mutable and list immutable
- Speed: Tuples are faster than lists
- Memory consumption: Lists comsume more memory tann tuples
- Built in functionality
- Error prone
- Usability



```python
# Immutable data types are faster, so that tuples are faster than lists
# We see from the example that tuple takes less time than liat
import time
L= list (range(100000000))
T= tuple(range(100000000))
start= time.time()
for i in L:
  1*5
print('List time', time.time()-start)
start= time.time()
for i in T:
  1*5
print('Tuple time', time.time()-start)
```

```python
# Which one between tuples and list consume more space/memory?
# Tuple takes less memory than list
# Example
import sys
L= list(range(1000))
T= tuple(range(1000))
print('List size', sys.getsizeof(L))
print('Tuple size', sys.getsizeof(T))
# Immutable data types take less memory than mutable list
```

```python
# Built in fuctionality
# List has many more built in functionalities than tuple
# Becuase Lsits mutable and have many coding functions whereas
# Tuple is not mutable so that many functions like edit, delete, addition, append don't work on tuple
```

```python
# Error prone attributes
# List id more error porn than tuple
# See how thr tuple is better than or less error porn than list
a= [1,2,3]
b=a
a.append(4)
print(a)
print(b)
# When we append(4), then the changes happened in both (a) and (b)
# Because List is mutable, it points at same memory location
# So List is risky becaus we don't change anyting in (b) but it changed
```

```python
# We check the same example for tuple
a= (1,2,3)
b=a
# We don't have append() function in tuple because tuple is immutable
# So we do
a= a + (4,)
print(a)
print(b)
# We see change in (a) but (b) is unchanged because tuple is immutable
# Less risk of error or change or wrong
```

# Special syntex
- Tuple unboxing/unpacking
- We can unpack a value of tuple and can store in variable

```python
# Upacking Tuple
a,b,c= (1,2,3)
print(a,b,c)
```

```python
# Example 2
a,b= (1,2,3)
print(a,b)
# Values in the tuple are more than varibles, only a and b but value 1,2,3
# We have to take equal numbers of variables and values in the tuple
# Make sure both sides are equal
# Else it will show an error
```

```python
# How to swap value in tuple
# Example
a= 1
b= 2
a,b= b,a
print(a,b)
```

```python
# When we want to extract oly first two values from a tuple but not all
# We do
a,b, *others= (1,2,3,4,5,6,7,8,9)
print(a,b)
# a=1 and b=2 here
print(others)
# Tuple unpacking done for all codes above
```

```python
# Zipping tuples
# Example
a= (1,2,3,4,5)
b= (6,7,8,9,10)
zip(a,b)
# print a zib object in the result
```

```python
# If we want to see zip() object value we can make a list or tuple function of zip()
# Zipping tuples
# Example
a= (1,2,3,4,5)
b= (6,7,8,9,10)
zip(a,b)
# print a zib object in the result
list(zip(a,b))
```

```python
# If we want to see zip() object value we can also make a tuple function of zip()
# Zipping tuples
# Example
a= (1,2,3,4,5)
b= (6,7,8,9,10)
zip(a,b)
# print a zib object in the result
tuple(zip(a,b))
```
