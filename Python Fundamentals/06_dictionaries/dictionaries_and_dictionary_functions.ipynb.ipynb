# Remove key-value pair from dictionary using
- pop
- popitem
- del
- clear

# pop and del work exactly same way


```python
# pop. on dict to remove a key-apir for existing dict
d13= {'name': 'Akter', 'age': 43, 3: 3, 'gender': 'male', 'weight': 78}
#d13.pop('gender') # it extract key-paired value
d13.pop('gender')
print(d13)
# pop removed 'gender' key and its value
```

```python
# popitem() function on dict to delete/remove only last key-value paired item
d13= {'name': 'Akter', 'age': 43, 3: 3, 'gender': 'male', 'weight': 78}
d13.popitem()
d13.popitem()
print(d13)
# popitem deletes/removes last key-value paired tem from dict everytime
```

```python
# del function, either can delete whole dict or specific key-value pair
# del
d14= {'name': 'Akter', 'age': 43, 3: 3, 'gender': 'male', 'weight': 78}
del d14['name']
print(d14)
```

```python
# del function, either can delete whole dict or specific key-value pair
# del
d14= {'name': 'Akter', 'age': 43, 3: 3, 'gender': 'male', 'weight': 78}
del d14
print(d14)
```

```python
# clear() on dict to make the whole dict to an empty dictionary
d15= {'name': 'Akter', 'age': 43, 3: 3, 'gender': 'male', 'weight': 78}
d15.clear()
print(d15)

# print an empty dict
```

# Dictionary operations
* There are two operations in dict

- Membership( It works on keys only)
- Iteration( Loops)

```python
# The mmebership operation works only on key

print(s)
'Akter' in s
# Membership operator looks for any key as 'Akter' in dict
# But there is no key as 'Akter' so it shows 'False'
# In dictionary everything is about key and doesn't care about value
```

```python
# The mmebership operation works on key or look for key in dict
print(s)
'name' in s
# Membership operator looks for any key as 'name' in dict
# There is a key as 'name' in the dict so it shows 'True'
# In dictionary everything is about key and doesn't care about value
```

```python
# How loops/iterations work on dict
# In dict everything is sbout keys
d13= {'name': 'Akter', 'age': 43, 3: 3, 'gender': 'male', 'weight': 78}
for i in d13:
 print(i)

 # Loops work on only keys
 # It prints only keys but ni vlaues
```

# Dictionary comprehension
- Creating dict in just one line using the given expression
# {key: value for vars in iterable}Python-dictionary-comprehension.webp


```python
# print first 10 numbers and their suqares
{i:i**2 for i in range(1,11)}

# i= the key, i**2(square of values)
```

```python
# uising existing distionary for dictionary comrehension
distances_kms= {'dhaka':250, 'ctg':200, 'sylhet': 100}
{key:value*0.62 for (key,value) in distances_kms.items()}
# we wnat to cahnge the distance kms to miles
# Here we keep the keys same and multiplying value(kms) with 0.6 to convert to miles
```

```python
# Using zip() on given dict
# We use keys as weekdays and value as temperatures for those weekdays
days= ['sunday', 'monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday']
temp_c= [30.5,32.4,31.5,30.7,32.33,31.22,30.9]
{i:j for i,j in zip(days,temp_c)}
```

```python
# Using if condition on dict

products= {'phone':10, 'laptop':5, 'charger':30, 'table':0}
{key:value for (key,value) in products.items() if value>0}
# We are checking the value of the given values are greater than 0 or not
# In the result key 'table' has vanished because no 'tablet' has been delivered
```
