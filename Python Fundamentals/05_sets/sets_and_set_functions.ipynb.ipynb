# Sets(data type)

 * A set is a data type used to store unique elements.
 * Sets have curly braces {}

1. Unordered(tuples and List matter order)
2. Unindexed(tuples and List maintained index)
3. Set is immutable( cannot be changed)
4. Set is mutable-itself in a sense that we can only add or remove items from it.
5. Set does not allow any duplicates
6. Set cannot contain mutable data types
7. We cannot have set inside set or 2D, 3D etc

* Union
* Intersection
* Symmetric difference and so on

```python
# How to access in an item in set or how to petch/extract an item from set
# For extracting an item we need positive and negative indexing
s1= {1,2,3,4,5}
s1[0] # positive idexing
s1[-1] # Negative indexing
# We have result an 'error' bcause set is unindexed means
# Set doesn't follw order so that the items in sets aren't ordered by indexing
# We cannot say [0]=1, [1]=2, [2]=3 and so on
# Mian reason that sets do not follow order, means set is unordered
# So we aren't able to execute positive and negative indexing in sets
```

```python
# Slicing is also not possible in sets
s1= {1,2,3,4,5}
s1[0:3] #Positive
s1[-3:-2] # Negative indexing

# Once items assigned in set, we cannot extract then by slicing/indexing because
# The sets are unordered
```

# Editing items in set

- Editing/replacing items in set isn't allowed


```python
# Editing/Replacing items in set
s1= {1,2,3,4}
s1[1]= 5
# Result error because 'idexing' doesn't work in set
# so that Editing/replacing isn't work
# Once the items assigned in set, we cannot extract/edit/replace them
```

```python
# However we can use 'add' function to add only one single new item in the set
# New item adiition in set
s2= {1,2,3,4}
s2.add(5)
print(s2)
# We have added new item 5 to the set
# We only able to add one single item individually and cannot add multiple items at once
# We also don't know in which position we add an item in the set and it is decided by hash(#) function works behind set
# So we have no controll where to add new items and only hash(#) that decides
```

```python
# We use 'update' function to add multiple items in the set
s2= {1,2,3,4}
s2.update([5,6,7])# we use mutable lsit [5,6,7] to update the set
print(s2)
# s2 has been updated
# Add and updates work as append() and extend() functions
```

# Deleting an item from set by
- del( to delete whole set)
- discard(to delete individual item)
- remove
- pop
- clear

```python
# To delete all items or  whole set, we use delete function
s1= {1,2,3,4}
print(s1) # Print s1 successfully
del s1
print(s1) # Delete s1 successfully so that error occurs
# All items of set s1 have been deleted so print nothing and leavs error
# We cannot delete individual items from set like s1[0]=1 by idexing because in set indexing doesn't work
# No idexing/slicing in set allowed
```

```python
# We use discard function to delete an induvidual item from set
s1= {1,2,3,4}
s1.discard(5)
print(s1) # 5 has been deleted

# If we want to delete an item which is not existing in the set we use discard also
s1.discard(10) # Show no result or error because item 10 doesn't exist in set s1
```

```python
# We use remove function to remove single items from the list
s1= {1,2,3,4,5}
print(s1)# print the set
s1.remove(5)# Remove 5
print(s1)

# remove is similar as discard but there is a main differnce between them
# If we want to remove a non existing item from the set, remove function shows error but
# Discard function doesn't show error in this case
s1.remove(7) # 7 does not exist in the set s1 and remove shows error
```

```python
# pop function remove only one item randomly from the set
s1= {1,2,3,4,5}
s1.pop()
# So pop removes single item but randomly
```

# Four major function update
- Update functions on sets

```python
# union-update function on sets
s1= {1,2,3,4,5,6}
s2= {5,6,7,8,9}
s1.union(s2)
s1.update(s2)
print(s1) # print union of s1 and s2
print(s2) # print only s2 and nothing changed
```

```python
# intersection-update function on sets
s1= {1,2,3,4,5,6}
s2= {5,6,7,8,9}
s1.intersection(s2)
s1.intersection_update(s2)
print(s1) # print the intesection of s1 and s2
print(s2) # print only s2 without any change
```

```python
# difference/difference update
s1= {1,2,3,4,5,6}
s2= {5,6,7,8,9}
s1.difference(s2)
s1.difference_update(s2)
print(s1) # print the elements of s1 which aren't in s2
print(s2) # Only print the s2 without any change
```

```python
# symmetric_difference/symmetric_difference_update
s1= {1,2,3,4,5,6}
s2= {5,6,7,8,9}
s1.symmetric_difference(s2) # print all elements of s1 and s2 without common elements in them
s1.symmetric_difference_update(s2) # Only print s2 without change
print(s1)
print(s2)
```

# Four more necessary functions
- isdisjoint sets
- issubset

```python
# isdisjoint of sets
# disjoint mean there are no common elemensts in the sets s1 and s2
s1= {1,2,3,4,5,6}
s2= {5,6,7,8,9}
s1.isdisjoint(s2)
# s1 and s2 are not disjoint because the have common items in them
# There are common elements in s1 and s2 and print (False)
```

```python
# isdisjoint of sets
s1= {1,2,3,4,5}
s2= {6,7,8,9}
s1.isdisjoint(s2)
# So s1 and s2 are disjoint becuase all the elemenst in them are different
# There are no common elements between s1 ans s2 so result is (True)
```

```python
# issubset of sets(True)
s1= {1,2,3,4,5,6,7,8}
s2= {1,2,3,4}
s2.issubset(s1)
# s2 has its all elements in s1
# So s2 is a subset of s1
```

```python
# issubset of sets (False)
s1= {1,2,3,4,5,6,7,8}
s2= {1,2,3,4}
s1.issubset(s2)
# s2 doesn't contain all elements of s1
# So s1 is not a subset of s2
```

```python
# issuperset(Superset)
s1= {1,2,3,4,5,6,7,8}
s2= {1,2,3,4}
s1.issuperset(s2)
# superset means s1 contains all the element of s2
# So s1 is a superset
```

```python
# copy of sets
s1= {1,2,3,4,}
s2= s1.copy()
print(s1)
print(s2)
# s2 just coppied s1 and result is same for s2
```

# Forzen set
- frozen set is immutable in python(whereas set is itself is mutsble)
- Once we have created the frozen set we cannnot add or delete any elements from frozen set

* All other fuctions like union, intersection, symmetric, differences work on frozen set
* Because we do opeation on sets to read the values for the set but we do not change the sets(do not change elements in the sets, we only read them)

# So all functions should work on frozen set

- When we want to make an app where we only want to run read only operations we use frozenset
- Don't want to change anything

```python
# frozen set
fs= frozenset((1,2,3,4,5))
fs
# We can pass any itterable object like list, tuple, set and so on in a frozen set
# Once frozen set created, we cannot add or delete items for it
```

```python
# We do some necessary functions on frozen set even frozen set is immutable
# We do not change anything inside set, we just compare two frozen sets
fs1= frozenset([1,2,3,4])
fs2= frozenset([3,4,5])
fs1|fs2

#Union(|)/set1.union(set2)
```

```python
# set1.union(set2) also union
fs1= frozenset([1,2,3,4])
fs2= frozenset([3,4,5])
fs1.union(fs2)
```

```python
# Intersection on froenset
fs1= frozenset([1,2,3,4])
fs2= frozenset([3,4,5])
fs1&fs2
```

```python
# difference on frozenset
fs1= frozenset([1,2,3,4])
fs2= frozenset([3,4,5])
fs1-fs2
```

```python
# Symmetric differences(^) on frozenset
fs1= frozenset([1,2,3,4])
fs2= frozenset([3,4,5])
fs1^fs2
```

# Comrehension make thing easy of writing code in a very different way for calculating things just in a single code insted writing many lines code
